```python
# importing csv from my google drive
file_path = '/content/drive/MyDrive/Standards/USCGsmaller.txt'
```

```python
# Open and read the content of the file

with open(file_path, 'r') as file:

    text = file.read()



# Now, let's proceed to calculate the most common words and their frequencies

from collections import Counter

import re



# Clean and split the text into words

words = re.findall(r'\b\w+\b', text.lower())



# Count the frequency of each word

word_counts = Counter(words)



# Extract the most common words

most_common_words = word_counts.most_common(10)



# Prepare to visualize the word frequency distribution

import matplotlib.pyplot as plt



# Get the 30 most common words for a better visualization perspective

top_words = word_counts.most_common(30)

words, counts = zip(*top_words)



# Plotting

plt.figure(figsize=(10, 8))

plt.barh(words, counts, color='skyblue')

plt.xlabel('Frequency')

plt.ylabel('Words')

plt.title('Top 30 Most Common Words in the Text')

plt.gca().invert_yaxis()

plt.show()



most_common_words, top_words
```

```python
from nltk import bigrams

# Generate bigrams from the list of words
bigram_list = list(bigrams(words))

# Count the frequency of each bigram
bigram_counts = Counter(bigram_list)

# Extract the most common bigrams
most_common_bigrams = bigram_counts.most_common(10)

# Visualize the top 10 most common bigrams
bigram_words, bigram_freqs = zip(*most_common_bigrams)
bigram_labels = [' '.join(bigram) for bigram in bigram_words]

plt.figure(figsize=(10, 8))
plt.barh(bigram_labels, bigram_freqs, color='lightcoral')
plt.xlabel('Frequency')
plt.ylabel('Bigrams')
plt.title('Top 10 Most Common Bigrams in the Text')
plt.gca().invert_yaxis()
plt.show()

most_common_bigrams, bigram_labels
```
